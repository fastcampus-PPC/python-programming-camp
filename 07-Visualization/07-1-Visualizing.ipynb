# 1) Matplotlib 기초 실습

주피터 노트북에서 그래프를 그릴 때 선언해줘야 하는 명령어

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
```

plot의 default 상황에서 기준 데이터는 y임. y축의 범위는 [1,2,3,4] 이므로 1.0 ~ 4.0의 범위를 갖는 그래프가 생성됨.
x축의 데이터가 없을 경우, y축의 데이터에 기반하여 동일한 모양으로 추정해주는데, 0부터 시작하기 때문에 0.0 ~ 3.0의 범위를 가짐.

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

반면, 자료를 순서대로 기입하여 명시한 경우는 위와 같은 그래프가 생성됨.

```python
plt.plot([1,2,3,4], [1,4,9,16], 'bo-.') # 세번째 파라미터는 plot의 스타일과 관련된 파라미터.
plt.axis([0, 6, 0, 20])
plt.show()
```

## Line Property 컨트롤

```python
import numpy as np

# numpy를 이용한 데이터 생성
t = np.arange(0., 5., 0.2)

# 파라미터를 연속적으로 3개 단위로 입력 가능
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

plot 함수 안에 파라미터를 연속적으로 입력해서 다양한 그래프 생성 가능.
그래프 스타일 파라미터로 다르게 나타내는 것도 가능.

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b", 
         lw=5, ls="-.", marker="o", ms=10, mec="g", mew=5, mfc="r")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.axis([0, 50, -10, 30])
plt.show()
```

다양한 파라미터로 세부적인 디자인 조절이 가능. 자세한 내용은 pdf 슬라이드 참고.

```python
plt.plot(t, t**2, linewidth=2.0)
```

```python
line, = plt.plot(t, t**2, '-')
line.set_antialiased(False) # 안티 앨리어싱 해제
```

```python
lines = plt.plot(t, t**2, t+1, t**2 + 1)
plt.setp(lines, color='r', linewidth=2.0)
plt.setp(lines, 'color', 'b', 'linewidth', 2.0)
```

setp 함수를 통해 이미 그려진 라인 스타일에 덮어쓰기 가능. 함수 파라미터를 비워둘 경우, 이미 그려진 그래프의 정보를 가져옴.

```python
plt.setp(lines)
```

## 화면 분할과 그리드 기능

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211) # 2X1 행렬의 첫번째
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212) # 2X1 행렬의 두번째
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

여러 개의 그래프를 선언해놓고, show 함수를 사용하면 한번에 비교가 가능

```python
plt.figure(1) # 그림 번호
plt.subplot(211) # subplot : 그리드 번호
plt.plot([1, 2, 3])
plt.subplot(212)
plt.plot([4, 5, 6])

plt.figure(2)
plt.plot([4, 5, 6])
plt.show()
```

subplot 그리드 예시
- 111 : 1x1그리드에 첫 번째 subplot
- 234 : 2x3그리드에 네 번째 subplot
- nml : nxm그리드에 l번째 subplot

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
print(x)

# 히스토그램 그리기
plt.hist(x, 50, normed=True, facecolor='g', alpha=0.75) # x 데이터를 50구간으로 분할

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
import matplotlib.mlab as mlab

n, bins, patches = plt.hist(x, 50, normed=True, facecolor='g', alpha=0.75)
plt.plot(x)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

# add a 'best fit' line
y = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, y, 'r--')
plt.show()
```

## 범례 추가하기

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, ls="--", label="cosine") # 라벨을 붙여줬지만 legend 함수가 없으므로 범례는 보이지 않음
plt.plot(X, S, ls=":", label="sine")
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2) # 범례 설정하는 부분
plt.show()
```

## Ticks, 내부그리드 설정하기

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.grid(True)
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.grid(False)
plt.show()
```

default의 경우, 알아서 ticks를 설정해 주지만 이렇게 수동으로 설정이 가능.

## pie chart 그리기

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# 시각화할 데이터
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
fracs1 = [15, 30, 45, 10]
fracs2 = [35, 10, 30, 25]

explode = (0, 0.25, 0, 0)

the_grid = GridSpec(2, 2)
plt.subplot(the_grid[1, 0], aspect=1)
plt.pie(fracs1, labels=labels, autopct='%1.1f%%', shadow=True)
plt.subplot(the_grid[0, 1], aspect=1)
plt.pie(fracs2, explode=explode, labels=labels, autopct='%.0f%%', shadow=True)

plt.show()
```

## 예제 정답

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'r--')
plt.title('Two subplots')
plt.ylabel('plot1 y')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'b--')
plt.xlabel('plot2 x')
plt.ylabel('plot2 y')

plt.show()
```
